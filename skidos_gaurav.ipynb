```python
#5 Calculate the retention of subscribers MoM (month over month) and WoW (week over week).
active_subs_jan = df[(df['ExpireDate'] >= '2022-01-01') & (df['ExpireDate'] <= '2022-01-31')]['IdUser'].nunique()
active_subs_jan
```

```python
active_subs_feb = df[(df['ExpireDate'] >= '2022-02-01') & (df['ExpireDate'] <= '2022-02-28')]['IdUser'].nunique()
active_subs_feb
```

```python
# MoM retention rate:
mom_retention = (active_subs_feb / active_subs_jan) * 100
mom_retention
```

```python
active_subs_curr_week = df[(df['ExpireDate'] >= '2022-02-08') & (df['ExpireDate'] <= '2022-02-14')]['IdUser'].nunique()
active_subs_curr_week
```

```python
active_subs_prev_week = df[(df['ExpireDate'] >= '2022-02-01') & (df['ExpireDate'] <= '2022-02-07')]['IdUser'].nunique()
active_subs_prev_week
```

```python
wow_retention = (active_subs_curr_week / active_subs_prev_week) * 100
wow_retention
```

```python
# Top-5 products for trials & subscription MoM (month over month)

# Add columns for month of subscription and month of expire
df['SubscriptionMonth'] = df['SubscriptionDate'].dt.to_period('M')
df['SubscriptionMonth']
df['ExpireMonth'] = df['ExpireDate'].dt.to_period('M')
df['ExpireMonth']
```

```python
# Filter for data from Jan 1, 2022 to Jun 30, 2022
ndf = df[(df['SubscriptionDate'] >= '2022-01-01') & (df['SubscriptionDate'] <= '2022-06-30')]
ndf
```

```python
# Group the data by product and month of subscription, and count the number of trials and subscriptions
grouped = ndf.groupby(['IdProduct', 'SubscriptionMonth', 'Lifetime']).agg({'Id': 'count'}).reset_index()
grouped['Type'] = 'Trial'
grouped.loc[grouped['Lifetime'] >= 8, 'Type'] = 'Subscription'
grouped = grouped.groupby(['IdProduct', 'SubscriptionMonth', 'Type']).agg({'Id': 'sum'}).reset_index()
grouped
```

```python
# Group the data by product and month of expire, and count the number of subscriptions
expire_grouped = ndf.groupby(['IdProduct', 'ExpireMonth']).agg({'Id': 'count'}).reset_index()
expire_grouped = expire_grouped.rename(columns={'ExpireMonth': 'SubscriptionMonth', 'Id': 'Subscriptions'})
expire_grouped['Type'] = 'Subscription'
```

```python
# Merge the two groups of data and calculate the MoM change
merged = pd.concat([grouped, expire_grouped], ignore_index=True)
merged = merged.groupby(['SubscriptionMonth', 'Type', 'IdProduct']).agg({'Id': 'sum'}).reset_index()
merged['MoM Change'] = merged.groupby(['Type', 'IdProduct'])['Id'].pct_change()
merged = merged.dropna()
merged
```

```python
#7 Get the Top-5 products for trials MoM
trials_mom = merged[merged['Type'] == 'Trial']
trials_mom = trials_mom.sort_values(['SubscriptionMonth', 'MoM Change'], ascending=[False, False])
trials_mom = trials_mom.groupby('SubscriptionMonth').head(5)
trials_mom
```

```python
#7 Get the Top-5 products for subscriptions MoM
subscriptions_mom = merged[merged['Type'] == 'Subscription']
subscriptions_mom = subscriptions_mom.sort_values(['SubscriptionMonth', 'MoM Change'], ascending=[False, False])
subscriptions_mom = subscriptions_mom.groupby('SubscriptionMonth').head(5)
subscriptions_mom
```
